```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Data.csv", header = None, names = ['N', 'Merge Sort', 'Bubble Sort'])
X = dataset.iloc[:, 0].values
y1 = dataset.iloc[:, 1].values
y2 = dataset.iloc[:, 2].values
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(np.reshape(X, (len(X), 1)))[:, [1, 2]]
```

```python
from sklearn.linear_model import LinearRegression
merge_reg = LinearRegression()
bubble_reg = LinearRegression()
merge_reg.fit(X_poly, y1)
bubble_reg.fit(X_poly, y2)
```

```python
X_plot = np.arange(min(X), max(X), 0.1)
X_plot = np.reshape(X_plot, (len(X_plot), 1))
plt.figure(figsize = (16, 9))
plt.plot(X_plot, merge_reg.predict(poly_reg.fit_transform(X_plot)[:, [1, 2]]), color = 'blue')
plt.plot(X_plot, bubble_reg.predict(poly_reg.fit_transform(X_plot)[:, [1, 2]]), color = 'red')
```

```python
from sklearn.metrics import r2_score
print(r2_score(y1, merge_reg.predict(X_poly)))
print(r2_score(y2, bubble_reg.predict(X_poly)))
```
